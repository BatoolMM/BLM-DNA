## Building Missing residues in BLM structure (4O3M)

```python
from modeller import *
```

```python
code = 'BLM'
e = Environ()
m = Model(e, file=code)
aln = Alignment(e)
aln.append_model(m, align_codes=code)
aln.write(file=code+'.seq')
```

### Use this file to create an alignment file by adding missing residues manually\
1- Copy the text twice in the file 
2- Change the name of the file to `.ali`
3- Add dashes in place of missing residues as you can see in the PyMol seq viewer
4- Add missing residues to the second part of the seq

```python
!cat alignment.ali
```

```python
from modeller import *
from modeller.automodel import *    # Load the AutoModel class

log.verbose()
env = Environ()

# directories for input atom files
env.io.atom_files_directory = ['.', '../atom_files']

a = LoopModel(env, alnfile = 'alignment.ali',
              knowns = 'BLM_missing', sequence = '4o3m')
a.starting_model= 1
a.ending_model  = 1

a.loop.starting_model = 1
a.loop.ending_model   = 2
a.loop.md_level       = refine.very_slow

a.make()
```
